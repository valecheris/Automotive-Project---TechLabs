```python
vwc = vw.drop_duplicates(keep = 'first')
vwc.info()
```

```python
sns.pairplot(vwc, hue ='fuelType')
```

```python
sns.lmplot(x='mileage', y='price', data = vwc)
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.ylim(0,70000)
```

```python
sns.lmplot(x='mileage', y='price', col ='fuelType', data = vwc)
plt.ylim(0, 60000)
```

```python
sns.stripplot(y='price',  data = vwc)
```

```python
sns.relplot(x='mpg', y='price', data = vwc)
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='transmission', y ='price', data = vwc)

plt.subplot(1,2,2)
sns.boxplot(x='engineSize', y='price', data = vwc)
plt.tight_layout()
```

```python
sns.boxplot(x='fuelType', y = 'price', data = vwc)
```

```python
pearsoncorr = vwc.corr(method = 'pearson')
```

```python
sns.heatmap(pearsoncorr,  xticklabels=pearsoncorr.columns,
            yticklabels=pearsoncorr.columns,
            cmap='RdBu_r',
            annot=True,
            linewidth=0.5)
```
